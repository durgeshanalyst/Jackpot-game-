```python
import random
def jackpot_game():
  jackpot=random.randint(1,100)

  guess=int(input('guess your number: '))
  count=1
  while jackpot != guess:

    if guess > jackpot:
      print('guess lower')
      guess=int(input('guess your number: '))
      count+=1

    else:
      print('Guess higher')
      guess = int(input('guess your number: '))
      count+=1

  print(f' Exactly it was {jackpot},but you took {count} attempt')
jackpot_game()
ask_for_play=input('would you like to play again: y/n ')
if ask_for_play=='Y' or ask_for_play=='y':
  jackpot_game()
else:
  print('Thank you for your time !!')

```
